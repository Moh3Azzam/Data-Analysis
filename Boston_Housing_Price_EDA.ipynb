## Import and Install

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.datasets import load_boston

%matplotlib inline
```

## Load Data

```python
boston = load_boston()
```

```python
boston.feature_names
```

```python
print(boston.DESCR)
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df.head()
```

```python
df['PRICE'] = boston.target
```

```python
df.head()
```

```python
sns.pairplot(df.iloc[:,0:7],diag_kind='kde')
```

```python
sns.pairplot(df.iloc[:,7:14],diag_kind='kde')
```

```python
n_cols = len(df.columns)/2
n_rows = 2
columns = df.columns

plt.figure(figsize=(15,6))
sns.set_style('darkgrid')

for i in range(len(df.columns)):
    plt.subplot(n_rows,n_cols,i+1) 
    sns.histplot(df[columns[i]], kde=True, color='tomato',element='poly')

plt.tight_layout()
```

```python
n_cols = len(df.columns)/2
n_rows = 2
columns = df.columns

plt.figure(figsize=(15,6))
sns.set_style('darkgrid')

for i in range(len(df.columns)):
    plt.subplot(n_rows,n_cols,i+1) 
    sns.scatterplot(data=df, x='PRICE', y=df[columns[i]], color='tomato')

plt.tight_layout()
```

```python
price_corr = round(df[df.columns[:]].corr()['PRICE'],2)
corr = pd.DataFrame(price_corr.values, columns=['Corr'], index=[price_corr.index])
top_corr = corr[(corr['Corr'] >= 0.5) | (corr['Corr'] <= -0.5)]
top_corr
```

```python
sns.pairplot(df.loc[:, ['PRICE', 'RM', 'PTRATIO', 'LSTAT']], corner=True, kind='reg',
            plot_kws={'line_kws':{'color':'blue'},
                           'scatter_kws': {'alpha': 0.5,
                                           'color': 'tomato'}},
            diag_kws=dict(color='tomato'))
```

```python
sns.heatmap(df.loc[:, ['PRICE', 'RM', 'PTRATIO', 'LSTAT']].corr(), annot=True, annot_kws={'fontsize':20}, cmap = 'Reds')
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(data=df, x='CHAS', y='PRICE', width=0.5);
```

```python
palette = sns.color_palette()
g = sns.pairplot( data=df, vars=['PRICE', 'RM', 'PTRATIO', 'LSTAT'], hue='CHAS', 
                 height=3, palette=[palette[2],palette[3]], 
                 )
g.map_offdiag(sns.scatterplot,  alpha=0.2)
```

```python
sns.set_style('darkgrid')
sns.jointplot(data=df, x='PRICE', y='RM', kind='hex', color='m');
sns.jointplot(data=df, x='PRICE', y='RM', kind='kde');
sns.jointplot(data=df, x='PRICE', y='RM');
```

```python
plt.figure(figsize=(15,8))
sns.regplot(data=df, x='PRICE', y='RM', )
plt.xlim(df['PRICE'].min()-1, df['PRICE'].max()+1)
```

```python
plt.figure(figsize=(15,8))
sns.regplot(data=df, x='PRICE', y='RM', robust=True)
plt.xlim(df['PRICE'].min()-1, df['PRICE'].max()+1)
```

```python
sns.lmplot(data=df, x='PRICE', y='RM', hue='CHAS', height=6, aspect=2)
plt.xlim(df['PRICE'].min()-1, df['PRICE'].max()+1)
```

```python
sns.lmplot(data=df, x='PRICE', y='RM', col='CHAS', height=6, aspect=2,)
plt.xlim(df['PRICE'].min(), df['PRICE'].max()+1)
```

```python
sns.lmplot(data=df, y='PRICE', x='CHAS', height=6, aspect=2,x_estimator=np.mean)
plt.ylim(df['PRICE'].min(), df['PRICE'].max()+1)
plt.xlim(-0.25, 1.25);
```
