```python
sns.set_style('darkgrid')
sns.relplot(x='sepal_length', y='sepal_width', data=df, hue='species' ,
            height=4, aspect=2)
```

**Observation**
- Using sepal_length and sepal_width features, we can distinguish Setosa flowers from others.
- Seperating Versicolor from Viginica is much harder as they have considerable overlap.

```python
sns.set_style('darkgrid')
sns.relplot(x='petal_length', y='petal_width', data=df, hue='species' ,
            height=4, aspect=2)
```

**Observation**
- Using petal_length and petal_width features, we can distinguish Setosa flowers from others.
- if petal_length < 5 and > 2 and petal_width <= 1.5 and >= 1 then it Versicolor

```python
sns.set_style('whitegrid')
sns.displot(data=df, x='petal_length' ,kde=True, hue='species')
```

**Observation**
- we can use petal_length (*PL*) to make classification 
- if PL $\le 2$ then **sentosa**
- if PL $ \le 4.5 $ and $\ge 2.5 $ then **versicolor**
- if PL $\le 7$ and $>4.5$ then **virginica**
